```python
import re

def generate_citation(text_info):
    # Extract the author's last name
    match = re.search(r'(\n\n[^\.]*\.)\n\nAuthor information', text_info, re.DOTALL)
    sentence = match.group(1).replace('\n', ' ')
    author_lastname = sentence.split()[0]
    
    # Extract the year of publication
    year = re.search(r'\. (\d{4}) ', text_info).group(1)
    
    # Extract the journal abbreviation
    journal_abbr = re.search(r'\. ([^\.]*\.)', text_info).group(1).strip(".")
    
    # Extract DOI number for citation
    match = re.search(r'\nDOI: (.*)\nPMID', text_info)
    match2 = re.search(r'\nDOI: (.*)\nPMCID', text_info)
    if match:
        doi_number = match.group(1).strip()
    elif match2:
        doi_number = match2.group(1).strip()
    else:
        doi_number = 0.0
    
    # Extract PMID
    match = re.search(r"PMID: (\d+)", text_info)
    if match:
        pmid_number = match.group(1)
    else:
        pmid_number = 0.0
    
    # Construct citation
    citation = f"{author_lastname} ({year}), {journal_abbr}., {doi_number}"
    
    return citation, pmid_number

citation_eg ="Jezierska (2014) Neurochem., 12491"

# Test with your specific string
text_info = "1. J Neurochem. 2014 Mar;128(5):741-51. doi: 10.1111/jnc.12491. Epub 2013 Nov 13.\n\nSCA14 mutation V138E leads to partly unfolded PKCŒ≥ associated with an exposed \nC-terminus, altered kinetics, phosphorylation and enhanced insolubilization.\n\nJezierska J(1), Goedhart J, Kampinga HH, Reits EA, Verbeek DS.\n\nAuthor information:\n(1)Department of Genetics, University of Groningen, University Medical Center \nGroningen, Groningen, The Netherlands.\n\nThe protein kinase C Œ≥ (PKCŒ≥) undergoes multistep activation and participates in \nvarious cellular processes in Purkinje cells. Perturbations in its \nphosphorylation state, conformation or localization can disrupt kinase \nsignalling, such as in spinocerebellar ataxia type 14 (SCA14) that is caused by \nmissense mutations in PRKCG encoding for PKCŒ≥. We previously showed that SCA14 \nmutations enhance PKCŒ≥ membrane translocation upon stimulation owing to an \naltered protein conformation. As the faster translocation did not result in an \nincreased function, we examined how SCA14 mutations induce this altered \nconformation of PKCŒ≥ and what the consequences of this conformational change are \non PKCŒ≥ life cycle. Here, we show that SCA14-related PKCŒ≥-V138E exhibits an \nexposed C-terminus as shown by fluorescence resonance energy \ntransfer-fluorescence lifetime imaging microscopy in living cells, indicative of \nits partial unfolding. This conformational change was associated with faster \nphorbol 12-myristate 13-acetate-induced translocation and accumulation of fully \nphosphorylated PKCŒ≥ in the insoluble fraction, which could be rescued by \ncoexpressing PDK1 kinase that normally triggers PKCŒ≥ autophosphorylation. We \npropose that the SCA14 mutation V138E causes unfolding of the C1B domain and \nexposure of the C-terminus of the PKCŒ≥-V138E molecule, resulting in a decrease \nof functional kinase in the soluble fraction. Here, we show that the mutation \nV138E of the protein kinase C Œ≥ (PKCŒ≥) C1B domain (PKCŒ≥-V138E), which is \nimplicated in spinocerebellar ataxia type 14, exhibits a partially unfolded \nC-terminus. This leads to unusually fast phorbol 12-myristate 13-acetate-induced \nmembrane translocation and accumulation of phosphorylated PKCŒ≥-V138E in the \ninsoluble fraction, causing loss of the functional kinase. In contrast to \ngeneral chaperones, coexpression of PKCŒ≥'s 'natural chaperone', PDK1 kinase, \ncould rescue the PKCŒ≥-V138E phenotype.\n\n¬© 2013 International Society for Neurochemistry.\n\nDOI: 10.1111/jnc.12491\nPMID: 24134140 [Indexed for MEDLINE] "

citation, pmid = generate_citation(text_info)
print(citation)
print(pmid)
```

```python
def extract_abstract(text_info):
    paragraph = re.search(r'Author information:(.*)DOI:', text_info, re.DOTALL).group(1)
    
    # Remove leading and trailing whitespace
    paragraph = paragraph.strip()
    
    return paragraph
    
```

```python
from typing import Dict, Set

data_dict: Dict[str, Dict[str, str]] = {}

# Assume we have lists of DOIs, abstracts, citations, and PMIDs
dois = ['doi1', 'doi2', 'doi3']
abstracts = ['abstract1', 'abstract2', 'abstract3']
citations = ['citation1', 'citation2', 'citation3']
pmids = ['pmid1', 'pmid2', 'pmid3']

# Use a for loop to add keys and values to the dictionary
for doi, abstract, citation, pmid in zip(dois, abstracts, citations, pmids):
    data_dict[doi] = {'abstract': abstract, 'citation': citation, 'pmid': pmid}

print(data_dict)
```
